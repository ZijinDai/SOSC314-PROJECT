## January 2025 — Policy Reversal and Federal Escalation

**January 22, 2025**  
The U.S. Department of Homeland Security (DHS) announced the rollback of an Obama-era directive that had restricted immigration enforcement in so-called *sensitive locations*, including hospitals, places of worship, schools, courtrooms, funerals, and weddings.

**January 24, 2025**  
Acting Secretary of Homeland Security **Benjamine Huffman** stated that individuals temporarily admitted into the United States under the Biden administration would now be subject to deportation.

**Early 2025**  
- The U.S. military was deployed in multiple states to assist ICE operations.  
- The administration invoked the **Alien Enemies Act**, accelerating deportations with minimal due process, including transfers of detainees to prisons in El Salvador.  
- Reports emerged of U.S. citizens being detained or deported, prompting legal challenges and widespread criticism from legal scholars and civil rights organizations.

During this period, ICE:
- Introduced **daily arrest quotas**  
- Pursued a strategy of **high-visibility public arrests**  

Federal courts later found evidence of **racial profiling**, with ICE accused of disproportionately targeting *“Hispanic-looking”* individuals.

---

## June 2025 — Nationwide Protests and Militarized Policing

**June 2025**  
Large-scale protests against ICE and mass deportation policies erupted nationwide, most prominently in **Los Angeles**, where the **California National Guard** was deployed with riot shields.

Legal and law enforcement experts warned that ICE’s increased reliance on **plainclothes arrests** contributed to a surge in **ICE impersonation crimes**, raising serious public safety concerns.

---

## July 2025 — Funding Expansion, Abuses, and Violent Incidents

**July 4, 2025**  
President **Donald Trump** signed the *One Big Beautiful Bill Act* into law, allocating **unprecedented funding to ICE**—exceeding that of any other federal law enforcement agency and surpassing the federal prison system.

Funding provisions included:
- Expansion of detention facilities  
- Intensified deportation operations  
- Large-scale agent recruitment  
  - \$50,000 signing bonus  
  - \$60,000 student loan forgiveness  

**July 4, 2025**  
A group attacked an ICE detention center in **Alvarado, Texas**, marking the first of several violent incidents targeting ICE facilities that year.

**July 21, 2025**  
*The Guardian* reported alleged abuses at ICE detention facilities in Florida, informally referred to as **“Alligator Alcatraz.”**  
Advocacy organizations documented:
- Deprivation of food and water  
- Denial of showers  
- Multiple deaths in ICE custody during the first months of Trump’s second term  

**By July 2025**  
Public opinion polls indicated **majority disapproval of ICE**, surpassing negative sentiment levels seen during the 2018 *“Abolish ICE”* movement.

---

## August–September 2025 — Polarization and Escalating Violence

**Late August 2025**  
A **Pew Research Center** report documented a sharply widening partisan divide:
- **72% of Republicans** viewed ICE favorably  
- **78% of Democrats** viewed ICE unfavorably  

**August 25, 2025**  
A man was arrested at a **Dallas ICE field office** after claiming to possess a bomb.

**September 24, 2025**  
A gunman fired on an ICE transport van near the Dallas field office:
- One detainee was killed at the scene  
- Two others were critically injured  
- The attacker later committed suicide  
- One injured victim subsequently died  

**September 2025**  
Protests against ICE and mass deportation continued at **Federal Plaza in Manhattan**, reflecting sustained nationwide mobilization.

---

## November–December 2025 — Institutional Shakeup and Intensified Raids

**By November 2025**  
At least **half of ICE’s senior leadership** was fired or reassigned, with many roles filled by **Border Patrol officials**.  
Media analysis suggested this restructuring reflected the administration’s push for **more aggressive and less targeted enforcement strategies**.

**Late November 2025**  
ICE launched a series of immigration raids in **New Orleans**, drawing criticism from local officials and prompting calls for improved transparency and interagency coordination.

**December 2025**  
- ICE detention reached **68,440 individuals**  
- Nearly **40,000 detainees** had no criminal record or only pending charges  

From **January to December 2025**:
- **328,000+ arrests**  
- **327,000 deportations**

---

## January 2026 — Minnesota Shooting and National Crisis

**January 7, 2026**  
An ICE agent shot and killed **Renee Nicole Good (37)** in her car in **Minneapolis, Minnesota**, during an enforcement-related incident involving members of the Somali community.

The shooting triggered:
- Immediate and widespread **anti-ICE protests**  
- President Trump’s threat to invoke the **Insurrection Act of 1807** to deploy federal troops to Minneapolis  

**January 7–18, 2026**  
The incident intensified national debate over:
- ICE’s legitimacy  
- Use of force  
- Role as a federal coercive institution  

This period marked a **critical escalation point** in both public and media discourse surrounding ICE.

```python
# Install all necessary packages
!pip install gensim pyLDAvis wordcloud -q

import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import nltk
import gensim
import pyLDAvis
import pyLDAvis.gensim_models
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
from gensim import corpora, models
from gensim.models import Phrases
from gensim.models.phrases import Phraser
from wordcloud import WordCloud

# Method 1: Using Google Drive (Recommended)
from google.colab import drive
drive.mount('/content/drive')
file_path = "/content/drive/MyDrive/CNN_Video_Final.csv"

# Alternative Method 2: Direct upload
# from google.colab import files
# uploaded = files.upload()
# file_name = list(uploaded.keys())[0]
# file_path = file_name

# Read the data
df = pd.read_csv(file_path)
df = df.dropna(subset=['transcript'])

# Download NLTK data
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')

# Preprocessing settings
stop_words = set(stopwords.words('english'))
custom_stops = {
    'nbsp', 'uh', 'um', 'know', 'just', 'going', 'like', 'said',
    've', 'dont', 'get', 'say', 'yeah', 'could', 'actually',
    'would', 'also', 'really', 'much', 'well', 'thats',
    'gtgt'  # Added gtgt as requested
}
stop_words.update(custom_stops)

def clean_and_tokenize(text):
    # Remove &nbsp; tags
    text = re.sub(r'&nbsp;', ' ', text)
    # Convert to lowercase, remove non-alphabetic characters
    text = re.sub(r'[^a-zA-Z\s]', '', text.lower())
    # Tokenize
    tokens = word_tokenize(text)
    # Filter stopwords and short words
    return [w for w in tokens if w not in stop_words and len(w) > 2]

# Apply preprocessing
df['processed_tokens'] = df['transcript'].apply(clean_and_tokenize)
processed_docs = df['processed_tokens'].tolist()

# Generate Bigrams
bigram = Phrases(processed_docs, min_count=5, threshold=10)
bigram_mod = Phraser(bigram)
processed_docs = [bigram_mod[doc] for doc in processed_docs]

# Count high-frequency words
all_words = [word for tokens in processed_docs for word in tokens]
word_freq = Counter(all_words).most_common(50)

print("Top 50 High-Frequency Words:")
for word, freq in word_freq:
    print(f"{word}: {freq}")

# Generate word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(word_freq))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud of Most Frequent Words')
plt.show()

# LDA Topic Modeling
# Create dictionary and corpus
dictionary = corpora.Dictionary(processed_docs)
corpus = [dictionary.doc2bow(text) for text in processed_docs]

# Train LDA model
num_topics = 5
lda_model = gensim.models.LdaModel(
    corpus=corpus,
    id2word=dictionary,
    num_topics=num_topics,
    passes=15,
    random_state=42
)

print("\nTopic Modeling Results:")
for idx, topic in lda_model.print_topics(-1):
    print(f"Topic {idx}: {topic}")

# Visualization
pyLDAvis.enable_notebook()
vis = pyLDAvis.gensim_models.prepare(lda_model, corpus, dictionary)
pyLDAvis.display(vis)

# Topic Weight Analysis
# Extract keywords and weights for each topic
topic_data = []
for i in range(num_topics):
    words = lda_model.show_topic(i, topn=10)
    for word, weight in words:
        topic_data.append([f"Topic {i}", word, round(weight, 4)])

df_topics = pd.DataFrame(topic_data, columns=['Topic', 'Word', 'Weight'])
print("\n### Detailed Keyword Weights for Each Topic ###")
print(df_topics.head(20))

# Calculate average proportion of topics in the corpus
def get_topic_distribution(corpus, model):
    topic_probs = []
    for bow in corpus:
        doc_topics = model.get_document_topics(bow, minimum_probability=0)
        topic_probs.append([prob for _, prob in doc_topics])
    return np.array(topic_probs)

probs_matrix = get_topic_distribution(corpus, lda_model)
mean_weights = probs_matrix.mean(axis=0)

print("\n### Average Weight of Each Topic in the Corpus ###")
for i, weight in enumerate(mean_weights):
    print(f"Topic {i}: {weight:.2%}")

# Plot topic distribution
plt.figure(figsize=(10, 6))
bars = plt.bar(range(num_topics), mean_weights, color='skyblue')
plt.xlabel('Topic ID')
plt.ylabel('Average Weight')
plt.title('Topic Distribution across Corpus')
plt.xticks(range(num_topics))

# Add value labels on bars
for bar, weight in zip(bars, mean_weights):
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{weight:.1%}', ha='center', va='bottom')

plt.show()
```
